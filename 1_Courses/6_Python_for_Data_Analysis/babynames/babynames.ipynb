```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

names2016 = pd.read_csv('names/yob2016.txt', names = ['name', 'sex', 'births'])
```

```python
years = range(1880, 2017)
pieces = []
columns = ['name', 'sex', 'births']
for year in years:
    path = 'names/yob%d.txt' % year
    frame = pd.read_csv(path, names = columns)
    frame['year'] = year
    pieces.append(frame)

names = pd.concat(pieces, ignore_index = True)
names
```

```python
total_births = pd.pivot_table(names, values = 'births', index = 'year', columns = 'sex', aggfunc = sum)
total_births
```

```python
total_births.plot(title = 'Total births by sex and year')
```

```python
def add_prop(group):
    births = group.births.astype(float)
    group['prop'] = births / births.sum()
    return group

names = names.groupby(['year', 'sex']).apply(add_prop)
names
```

```python
np.allclose(names.groupby(['year', 'sex']).prop.sum(), 1)
```

```python
def get_top1000(group):
    return group.sort_values(by = 'births', ascending = False)[:1000]

grouped = names.groupby('year')
top1000 = grouped.apply(get_top1000)
del top1000['year']
top1000
```

```python
boys = top1000[top1000.sex == 'M']
girls = top1000[top1000.sex == 'F']

total_births = top1000.pivot_table(values = 'births', index = 'year', columns = 'name', aggfunc = sum)
total_births
```

```python
subset = total_births[['John', 'Harry', 'Mary', 'Marilyn']]

subset.plot(subplots = True, figsize = (12, 12), grid = True, title = 'Number of births per year')
```

```python
table = top1000.pivot_table(values = 'prop', index = 'year', columns = 'sex', aggfunc = sum)

table.plot(title = 'Sum of table1000.prop by year and sex', figsize = (8, 5), grid = True, 
           yticks = np.linspace(0, 1.2, 13), xticks = range(1880, 2020, 10))
```

```python
def get_quantile_count(group, q = 0.5):
    group = group.sort_values(by = 'prop', ascending = False)
    return int(group.prop.cumsum().searchsorted(q) + 1)

diversity = top1000.groupby(['year', 'sex']).apply(get_quantile_count)
diversity = diversity.unstack('sex')
diversity.head()
```

```python
diversity.plot(grid = True, title = 'Number of popular names in top 50%')
```

```python
get_last_letter = lambda x: x[-1]
last_letters = names.name.map(get_last_letter)
last_letters.name = 'last_letter'
last_letters
```

```python
table = names.pivot_table(values = 'births', index = last_letters, columns = ['sex', 'year'], aggfunc = 'sum')
table
```

```python
subtable = table.reindex(columns = [1910, 1960, 2010], level = 'year')
subtable.head()
```

```python
subtable.sum()
```

```python
letter_prop = subtable / subtable.sum()
letter_prop
```

```python
fig, axes = plt.subplots(2, 1, figsize = (10, 8))
letter_prop['M'].plot(kind = 'bar', rot = 0, ax = axes[0], title = 'Male')
letter_prop['F'].plot(kind = 'bar', rot = 0, ax = axes[1], title = 'Female', legend = False)
```

```python
all_names = top1000.name.unique()
mask = np.array(['lesl' in x.lower() for x in all_names])
lesley_like = all_names[mask]
lesley_like
```

```python
filtered = top1000[top1000.name.isin(lesley_like)]
filtered.groupby('name').births.sum()
```

```python
table = filtered.pivot_table(values = 'births', index = 'year', columns = 'sex', aggfunc = 'sum')
table = table.div(table.sum(1), axis = 0)
table.tail()
```

```python
table.plot(figsize = (8, 5), style = {'M': 'k-', 'F': 'k--'})
```
