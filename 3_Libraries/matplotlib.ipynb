### ticks, labels, lengends

```python
ticks = ax.set_xticks([0, 250, 500, 750, 1000])
```

```python
labels = ax.set_xticklabels(['one', 'two', 'three', 'four', 'five'], rotation=30, fontsize='small')
```

```python
ax.set_title('Test plot')
```

```python
ax.set_xlabel('Stage')
```

```python
fig
```

```python
# lengends
fig, ax = plt.subplots(1, 1, figsize=(10, 6))
```

```python
ax.plot(randn(500).cumsum(), 'k', label='one')
ax.plot(randn(500).cumsum(), 'k--', label='two')
ax.plot(randn(500).cumsum(), 'k.', label='three')
```

```python
ax.legend(loc='best')
```

### plot in pandas

```python
# line plots
s = pd.Series(np.random.randn(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

```python
df = pd.DataFrame(np.random.randn(10, 4).cumsum(0), columns=['A', 'B', 'C', 'D'], index=np.arange(0, 100, 10))
df.plot()
```

```python
# bar plots
fig, axes = plt.subplots(2, 1)
```

```python
data = pd.Series(np.random.rand(16), index=list('abcdefghijklmnop'))
data.plot(kind='bar', ax=axes[0], color='k', alpha=0.7)
data.plot(kind='barh', ax=axes[1], color='r', alpha=0.3)
```

```python
fig
```

```python
df = pd.DataFrame(np.random.rand(6, 4), index=['one', 'two', 'three', 'four', 'five', 'six'], 
                  columns=pd.Index(['A', 'B', 'C', 'D'], name='Genus'))
df
```

```python
# histograms and density plots
comp1 = np.random.normal(0, 1, size=200)
comp2 = np.random.normal(10, 2, size=200)
```

```python
values = pd.Series(np.concatenate([comp1, comp2]))
```

```python
values.hist(bins=100, alpha=0.3, color='k', normed=True)
```

```python
values.plot(kind='kde', style='k--')
```
